```typescript
// Get puzzle input

import * as dotenv from "jsr:@std/dotenv";

const DAY: number = 1;
const INPUT_URL: string = `https://adventofcode.com/2024/day/${DAY}/input`
const input: string = await (await fetch(INPUT_URL, {
    headers: [["Cookie", `session=${dotenv.loadSync().SESSION_COOKIE}`]]
})).text()

input
```

```typescript
// Get test input

const test_input: string = `3   4
4   3
2   5
1   3
3   9
3   3`

test_input
```

```typescript
const solve = (_: string): number => {

    const locationIds = _.split("\n").map(line => line.split("   ")).reduce<number[][]>((acc, curr) => { acc[0].push(parseInt(curr[0])); acc[1].push(parseInt(curr[1])); return acc }, [[], []]).map(ids => ids.sort());

    
    
    let totalDistance: number = 0;
    
    for (let i = 0; i < locationIds[0].length; i++) {
        if(!isNaN(locationIds[0][i]) && !isNaN(locationIds[1][i])) {
            totalDistance += Math.abs(locationIds[0][i] - locationIds[1][i])
        }
    }
    
    return totalDistance;
}
```

```typescript
solve(test_input)
```

```typescript
solve(input)
```
